```python
# Description: This program attempts to predict the future price of Dogecoin
```

```python
# import some important libraries
import numpy as np
import pandas as pd
```

```python
# Load the dataset
df = pd.read_csv(r'C:\Users\Yash Kumar\Documents\Dogecoin price predict using Machine Learning\DOGE-USD.csv')
df = df.set_index(pd.DatetimeIndex(df['Date'].values))
df.head()
```

```python
df.info()
```

```python
print(pd.isnull(df).sum()) 
```

```python
df = df.dropna()
```

```python
print(pd.isnull(df).sum()) 
```

```python
# get the close price
df = df[['Close']]
df
```

```python
# create a variable to store the number of days into the future that we want to predict.
prediction_days = 1
# ctreate a new column prediction
df['prediction'] = df[['Close']].shift(-prediction_days)
# show the data
df
```

```python
# create the independent dataset (x)
X = np.array(df.drop(['prediction'], 1))
# remove the last n+1 rows of data where n is the predicton days
X = X[:len(df) - prediction_days - 1]
# print the data

print(X)
```

```python
# create the dependent data set (y)
y = np.array(df['prediction'])
# Get all of the y values except the last n+1 rows
y = y[:-prediction_days-1]
# show the data
print(y)
```

```python
X.shape
```

```python
print(pd.isnull(X).sum()) 
```

```python
print(pd.isnull(y).sum()) 
```

```python
y.shape
```

```python
# split the dataset into 80% training dataset and 20% testing dataset
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
print(pd.isnull(x_train).sum()) 
```

```python
# RandomForestRegressor for the model
from sklearn.ensemble import RandomForestRegressor
forest = RandomForestRegressor(n_estimators = 2, random_state = 587)
forest.fit(x_train, y_train)
print(forest.score(x_test, y_test))
```

```python
# show how close the predict values and actual values
prediction = forest.predict(x_test)
# print the predicted test values
print(prediction)
print()
# print the actual values
print(y_test)
```

```python
# Get the validation data for model
# create a variable to store all of the rows in the dataset except the last n rows
temp_df = df[:prediction_days]
# create to variable to store the independent price value
x_val = temp_df.tail(1)['Close'][0]
# show the data
print(x_val)
```

```python
prediction = forest.predict([[x_val]])
# print the price of Dogecoin for the next n days
print('The price of Dogecoin in', prediction_days, 'day(s) is predicted to be', prediction)
# print the actual value for the next n days
print('The actual price was', temp_df.tail(1)['prediction'][0])
```
